* Percentage and Count of Male Players

```python
#Types of gender
genders = purchase_data['Gender'].unique()
genders
```

```python
#Create a dataframe having unique players
unique_players_df = purchase_data.groupby('SN').first()

#Count male players
male_count = len(unique_players_df.loc[unique_players_df['Gender']=='Male'])

#Count female players
female_count = len(unique_players_df.loc[unique_players_df['Gender']=='Female'])

#count Other / Non-Disclosed players
other_count = len(unique_players_df.loc[unique_players_df['Gender']=='Other / Non-Disclosed'])

#total_genders should be equal to len(unique_players) = 576
total_genders = male_count+female_count+other_count
total_genders
```

```python
#Count male players
male_players = purchase_data.loc[purchase_data['Gender']=='Male']
unique_male_players = male_players['SN'].unique()
male_count = len(unique_male_players)
male_count
#count female players
female_players = purchase_data.loc[purchase_data['Gender']=='Female']
unique_female_players = female_players['SN'].unique()
female_count = len(unique_female_players)
female_count
#count Other / Non-Disclosed players
other_players = purchase_data.loc[purchase_data['Gender']=='Other / Non-Disclosed']
unique_other_players = other_players['SN'].unique()
other_count = len(unique_other_players)
other_count
#total_genders should be equal to len(unique_players) = 576
total_genders = male_count+female_count+other_count
```

```python
male_percent = male_count/total_genders*100
female_percent = female_count/total_genders*100
other_percent = other_count/total_genders*100
```

```python
#Create original demographics dataframe
gender_demographics = pd.DataFrame({'Gender':['Male','Female','Other / Non-Disclosed'],
                                   'Total Count':[male_count, female_count, other_count],
                                   'Percentage of Players':[male_percent, female_percent, other_percent]})
#Set index
gender_demographics = gender_demographics.set_index('Gender')
#Adjust formatting
gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map('{:.2f}%'.format)
gender_demographics
```

* Create a summary data frame to hold the results

```python
purchase_count = purchase_data.groupby(['Gender']).count()['Purchase ID']

avg_purchase_price = purchase_data.groupby(['Gender']).mean()['Price']

total_purchase_value = purchase_data.groupby(['Gender']).sum()['Price']

avg_total_buy_male = total_purchase_value.loc['Male']/male_count
avg_total_buy_female = total_purchase_value.loc['Female']/female_count
avg_total_buy_other = total_purchase_value.loc['Other / Non-Disclosed']/other_count
```

```python
gender_purchase_analysis = pd.DataFrame({'Purchase Count': purchase_count, 'Average Purchase Price': avg_purchase_price,
                                        'Total Purchase Value': total_purchase_value,
                                        'Avg Total Purchase per Person':[avg_total_buy_male, avg_total_buy_female,
                                                                        avg_total_buy_other]})
formatted_gender_purchase_df = pd.DataFrame()
formatted_gender_purchase_df['Purchase Count'] = gender_purchase_analysis['Purchase Count']
formatted_gender_purchase_df['Average Purchase Price'] = gender_purchase_analysis['Average Purchase Price'].map('${:.2f}'.format)
formatted_gender_purchase_df['Total Purchase Value'] = gender_purchase_analysis['Total Purchase Value'].map('${:.2f}'.format)
formatted_gender_purchase_df['Avg Total Purchase per Person'] = gender_purchase_analysis['Avg Total Purchase per Person'].map('${:.2f}'.format)
formatted_gender_purchase_df
```

```python
items_df.head()
```

```python
items_group = items_df.groupby(['Item ID','Item Name'])
```

```python
popular_purchase_count = items_group.count()['Price']
```
